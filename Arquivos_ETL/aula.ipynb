```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql
```

```python

def load_data_to_mysql(username, password, host, database, file_path, table_name):
    """
    Carrega dados de um arquivo txt para uma tabela no banco de dados MySQL.
    :param username: Nome de usuário para acessar o banco de dados.
    :param password: Senha para acessar o banco de dados.
    :param host: Host onde o banco de dados está hospedado.
    :param database: Nome do banco de dados.
    :param file_path: string com o caminho do arquivo a ser carregado.
    :param table_name: string com o nome da tabela onde os dados serão inseridos.
    """
    engine = create_engine(f'mysql+pymysql://{username}:{password}@{host}/{database}')

    # Ler o arquivo txt com dados separados por tabulações
    data = pd.read_csv(file_path, sep='\t', encoding='utf-8')

    # Insere os dados na tabela especificada
    data.to_sql(table_name, con=engine, if_exists='append', index=False)

    print(f"Dados importados com sucesso para a tabela {table_name}.")
```
